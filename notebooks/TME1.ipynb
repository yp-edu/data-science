We then transform degree coordinates into numerical coordinates to build a map


```python
import re

def dms2dd(degrees, minutes, seconds, direction):
    dd = float(degrees) + float(minutes)/60 + float(seconds)/(60*60);
    if direction in ('S', 'O'):
        dd *= -1
    return dd

def parse_dms(dms):
    #print(dms)
    parts = re.split('[^\d\w]+', dms)
    #print(parts)
    if (len(parts)<5):
      parts[3]=parts[2]
      parts[2]='00'
      parts.append('')
    #print(parts)
    lat = dms2dd(parts[0], parts[1], parts[2], parts[3])
    #lng = dms2dd(parts[4], parts[5], parts[6], parts[7])
    return lat

data['latitude2'] = data['latitude'].apply(parse_dms)
data['longitude2'] = data['longitude'].apply(parse_dms)
print(data)
```

The map can be obtain with the following code

```python
import geopandas as gpd
from pathlib import Path
import folium
from IPython.display import display

gdf = gpd.GeoDataFrame(
    data, geometry=gpd.points_from_xy(data.longitude2, data.latitude2))

Path("leaflet").mkdir(parents=True, exist_ok=True)

center = gdf[['latitude2', 'longitude2']].mean().values.tolist()
sw = gdf[['latitude2', 'longitude2']].min().values.tolist()
ne = gdf[['latitude2', 'longitude2']].max().values.tolist()

m = folium.Map(location = center, tiles='openstreetmap')

# I can add marker one by one on the map
for i in range(0,len(gdf)):
    folium.Marker([gdf.iloc[i]['latitude2'], gdf.iloc[i]['longitude2']], popup=gdf.iloc[i]['name']).add_to(m)

m.fit_bounds([sw, ne])
display(m)
#m.save("map.html")
#webbrowser.open("map.html")
```
